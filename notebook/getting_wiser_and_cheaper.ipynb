```python
# prompt: pandasで /content/drive/MyDrive/work/pycon_kyushu/data/model-eval-price.csv を読み込んでください。

import pandas as pd

df_price = pd.read_csv("/content/drive/MyDrive/work/pycon_kyushu/data/model-eval-price.csv")

# Print the DataFrame
print(df_price)
```

```python
# prompt: MMLUの列とMATHの列に % が入って文字列になっているので、小数に修正してください

df_price["MMLU(5shot)"] = df_price["MMLU(5shot)"].str.replace("%", "").astype(float) / 100
df_price["MATH"] = df_price["MATH"].str.replace("%", "").astype(float) / 100

# Print the updated DataFrame
print(df_price)
```

```python
model = 'gpt-4o-2024-05-13'
messages = []


system_prompt = '''
あなたはAIの権威でITトップ企業のCTOです。
データを見てアドバイスをください。
'''.strip()

messages.append({"role": "system", "content": system_prompt})

prompt = f"""
    次のデータはLLMの賢さ、APIの価格、リリース日の
    データです。
    データ: {df_price.to_markdown()}

    データから次のことを解説してください。

    - 賢さの指標について
    - 賢さと価格、リリース日の関係について分かること
    - これからLLMのAPIを使う人へのアドバイス
""".strip()

client = OpenAI(api_key=userdata.get('openai_api_key'))

def get_completion(prompt, model=model):
    messages.append({'role': 'user', 'content': prompt})
    res = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=0,
    )
    res = res.choices[0].message.content
    messages.append({'role': 'assistant', 'content': res})
    print(messages[-1]['content'])
```

```python
get_completion(prompt)
```

```python
get_completion('開発テストに用いるにはどのモデルが良いでしょうか？性能が良いわりに価格が安いモデルが良いです。コスト重視でお願いします。')
```
